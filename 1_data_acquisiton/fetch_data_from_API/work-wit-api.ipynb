```python
top_rated_movie_df = pd.DataFrame()
```

```python
for i in range(1, 571):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=7dc150ba53829db935fd6f7e970cc0dd&language=en-US&page={0}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = top_rated_movie_df.append(temp_df, ignore_index=True)
```

```python
df.shape
```

```python
for i in range(1, 501):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=7dc150ba53829db935fd6f7e970cc0dd&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)
```
