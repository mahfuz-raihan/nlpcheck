````python
import PyPDF2 as pdf
file = open('/media/mahfuz/Media/datasets/NLP_learn/data_acquisiton/work-with-pdf/NLP.pdf','rb')
````

#### Check the file type

````python
file
````

#### Read the PDF file

````python
pdf_reader = pdf.PdfReader(file)
````

````python
pdf_reader
````

#### Check the pdf reader insight

````python
help(pdf_reader)
````

#### if pdf id encrypted or not

````python
pdf_reader.is_encrypted
````

#### find pdf file pages

````python
len(pdf_reader.pages)
````

#### Extract text from the pdf file

````python
page2 = pdf_reader.pages[1]
````

````python
page2.extract_text()
````

````python
page3 = pdf_reader.pages[7]
page3.extract_text()
````

#### append write or Marge PDF's

````python
pdf_writer = pdf.PdfWriter()
````

````python
pdf_writer.add_page(page2)
pdf_writer.add_page(page3)
````

````python
output = open('pages.pdf', 'wb')
pdf_writer.write(output)
output.close()
````
